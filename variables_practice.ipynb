```python
# practicing variables

# variable stores data

x = 10
print("x=", x)

# x is the variable, 10 is the value(or string)

user1 = "FarahB"
_batch2 = "python"
print("Name:", user1)
print("Course:", _batch2)

# 1- Snake case
#All lowercase letters, words are separated by underscore
print("Snake case")
my_class = "python_batch2"
my_id = "0011"
print("Class:",my_class)
print("ID:", my_id)

# 2- Camel case
# First letter of first word is lowercase, first letter of each
# subsequent word is capitalized
print("Camel case")
myClass = "python_batch2"
myId = "0011"
print("Class:",myClass)
print("ID:", myId)

#3 Pascal case
# First letter of each word is capitalized
# no separator
print("Pascal case")
MyClass = "python_batch2"
MyId = "0011"
print("Class:",MyClass)
print("ID:", MyId)

```
